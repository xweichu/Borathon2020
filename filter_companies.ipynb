```python
# ENterprise Kaggle data
ent = pd.read_csv('companies_sorted.csv')
ent = pd.DataFrame(ent)
ent = ent.dropna()
ent
```

```python
# Filetring only the US as we don't have corona info by city for the whole world in one dataset 
us_data = ent[ent['country'] == 'united states']
us_data = pd.DataFrame(us_data)
us_data.reset_index() 
```

```python
us_data
```

```python
# covid data from https://www.kaggle.com/imdevskp/corona-virus-report?select=usa_county_wise.csv latest avaliable day
covid_us = pd.read_csv('datasets_515695_1387715_csse_covid_19_data_csse_covid_19_daily_reports_07-31-2020.csv')
covid_us = pd.DataFrame(covid_us)
covid_us = covid_us.dropna()
```

```python
covid_us = covid_us.drop(['FIPS', 'Last_Update', 'Lat', 'Long_'], axis = 1)
```

```python
covid_us
```

```python
# preparing the dataset for merging 
covid_us.columns = ['city', 'state', 'country', 'confirmed', 'deaths', 'recovered', 'active', 'locality',
                    'incidnece_rate', 'case_fatality_rate']
```

```python
covid_us['Combined_Key'] = covid_us['Combined_Key'].str.replace('US','united states')
covid_us['Combined_Key'] = covid_us['Combined_Key'].str.lower()
```

```python
# Merging the two datat sets on locality 
data = pd.merge(us_data, covid_us, how = 'inner', on = 'locality')
```

```python
data # 189639 rows
```

```python
# dropping not so useful columns 
data = data.drop(['Unnamed: 0', 'domain', 'year founded', 'locality', 'country_x'], axis=1) 
```

```python
# lists of the possible values, so that we can get to know the data and filter
industry = data['industry'].unique()
size_range = data['size range'].unique()
```

```python
# Companies Ranked by incidnece rate in the city they are in or registered at
data = data.sort_values(by = 'incidnece_rate', ascending = False)
```

```python
data.to_csv ('data.csv', index = False, encoding='utf-8')
```
